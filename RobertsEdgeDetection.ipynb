```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
x = np.array([
    [-1,0],[0,1]
])

y = np.array([
    [0,-1],
    [1,0]
])
```

```python
img = cv2.imread('img.jpg')
```

```python
cv2_imshow(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(img)
```

```python
gx = cv2.filter2D(img, -1, x)
```

```python
cv2_imshow(gx)
```

```python
gy = cv2.filter2D(img, -1, y)
```

```python
cv2_imshow(gy)
```

```python
image = np.sqrt(gx**2 + gy**2)
```

```python
cv2_imshow(image)
```

```python
im = (gx+gy)/2
cv2_imshow(im)
```
